## model architectures

many models have been proposed for estimating continuous word representations, including latent semantic analysis (lsa) and latent dirichlet allocation (lda). but the distributed representations learned by neural networks have shown better performance. the computational complexity of these models is defined as:

$$
o = e \times t \times q
$$

where:
- $e$ is the number of training epochs
- $t$ is the number of words in the training set
- $q$ is defined for each model architecture

### feedforward neural net language model 
the feedforward neural net language model consists of input, projection, hidden, and output layers. at the input layer, $n$ previous words are encoded using a 1-of-v coding, where $v$ is the vocabulary size. the projection layer has dimensionality $n \times d$. the computational complexity per training example is:

$$
q = n \times d + n \times d \times h + h \times v
$$

where:
- $d$ is the dimensionality of the word vectors
- $h$ is the size of the hidden layer
- $v$ is the size of the vocabulary

## new log-linear models
the paper introduces two new model architectures to learn distributed representations of words with reduced computational complexity:

1. **continuous bag-of-words model (cbow)**: the cbow model predicts the current word based on its context by averaging the word vectors in the context. the training complexity is:

$$
q = n \times d + d \times \log_2(v)
$$

2. **continuous skip-gram model**: the skip-gram model maximizes the classification of a word based on another word in the same sentence. the training complexity is:

$$
q = c \times (d + d \times \log_2(v))
$$

where $c$ is the maximum distance between words.

## results
the paper compares the quality of different word vector models by measuring their performance on semantic-syntactic word relationships. it shows that the new architectures (cbow and skip-gram) outperform the previous neural network models in accuracy while requiring significantly less computational cost.

## conclusion
the study shows that it is possible to train high-quality word vectors using simpler model architectures, such as cbow and skip-gram, which are computationally efficient and capable of learning from much larger datasets. this advancement could lead to new applications in nlp tasks like machine translation, information retrieval, and question-answering systems.
